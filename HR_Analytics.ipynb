```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings

warnings.filterwarnings('ignore')
```

Corelation Map

```python
fig, ax= plt.subplots(figsize=(20,20))
data_corr= data.drop(['Attrition','Department','BusinessTravel','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime'], axis=1)
sns.heatmap(data_corr.corr(),annot=True, linewidths=0.5, fmt='0.1f', ax=ax)
```

Countplots

```python
sns.countplot(x='OverTime', data=data,palette='deep')
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Over Time')
plt.show()
```

```python
sns.countplot(data=data, x="Education", palette='deep')
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Education Level')
plt.show()
```

```python
sns.countplot(x='MaritalStatus', hue='OverTime', data=data)
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Marital Status')
plt.show()

#getting the employee ids of the people who are the morottu singles
morattu_singles= data.loc[data['MaritalStatus'] == 'Single']
print("The employee ids of people who are single are as below:\n")
morattu_singles['EmployeeNumber']
```

```python
sns.countplot(x= 'JobRole',palette='deep', data=data)
fig= plt.gcf()
fig.set_size_inches(21,5)
plt.title('Job Role')
```

```python
sns.countplot(x='Gender', hue='OverTime' , data=data)
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Over Time Comparison wrt Gender')
```

```python
sns.countplot(data=data, x="EducationField", palette='deep')
fig= plt.gcf()
fig.set_size_inches(10,5)
plt.title('Education Field')
plt.show()
```

```python
sns.countplot(data=data, x="Department", palette='deep')
fig= plt.gcf()
fig.set_size_inches(6,5)
plt.title('Department')
plt.show()
```

```python
sns.catplot(data=data, x='OverTime', y='Age', kind='swarm', hue='OverTime')
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.title('Overtime vs Age Distribution')
```

```python
sns.countplot(data=data, x="BusinessTravel", palette='deep')
fig= plt.gcf()
fig.set_size_inches(6,5)
plt.title('Business Travel')
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(data=data, x='TotalWorkingYears',palette=sns.color_palette("cubehelix", n_colors=data['TotalWorkingYears'].nunique()))
plt.title('Working Years')
```

```python
sns.countplot(data= data, x='NumCompaniesWorked',palette='dark')
plt.title('Number of Companies Worked')
fig= plt.gcf()
fig.set_size_inches(5,5)
plt.show()
```

```python
fig, ax= plt.subplots(figsize=(18,5))
sns.countplot(data=data, x='DistanceFromHome', palette=sns.color_palette("cubehelix", n_colors=data['TotalWorkingYears'].nunique()))
plt.title('Distance From Home')
```
